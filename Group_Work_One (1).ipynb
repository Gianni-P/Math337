**Problem 18 from 1.1 of the Textbook; see page 12.** To find Equation (ii), I would rewrite the equation

$$
P(t+\Delta t) \approx P(t) + r\Delta t P(t) + k\Delta t
$$

using $t=n\Delta t$ so that 

$$
P(n\Delta t+\Delta t) \approx P(n\Delta t) + r\Delta t P(n\Delta t) + k\Delta t
$$
or 
$$
P((n+1)\Delta t) \approx P(n\Delta t) + r\Delta t P(n\Delta t) + k\Delta t
$$


Letting $P_{n}=P(n\Delta t)$ gives us the difference equation 

$$
P_{n+1} = P_{n} + r\Delta t P_{n} + k\Delta t = (1+r\Delta t)P_{n} + k\Delta t, ~ n\geq 0.
$$

A little bird in my brain then tells me that 
$$
P_{n} = (1+r\Delta t)^{n}\left( P_{0} + \frac{k}{r}\right) - \frac{k}{r}
$$
You do **NOT** need to prove this inductively.  That said, from this solution to the difference equation, you can derive Equation (ii) using the steps I showed in the notes.  

$\textbf Solution: $

We know: $ \frac{\delta P}{\delta t} = rP + k $
$$ $$
Using Taylor Expansion we have:
$$
P(t + \Delta t) = P(t) + P'(t)\Delta t + \mathcal{O}(\Delta t^2)
$$
When we isolate P'(t) we get:
$$
P'(t) = \frac{P(t + \Delta t) - P(t)}{\Delta t} - \mathcal{O}(\Delta t)
$$
Then substituting $ \frac{\delta P}{\delta t} = rP + k $ we get:
$$
rP + k = \frac{P(t + \Delta t) - P(t)}{\Delta t} - \mathcal{O}(\Delta t)
$$
Now isolating $P(t + \Delta t)$ and assuming $ \Delta t^2 \approx 0 $ we get:
$$
P(t + \Delta t) = P(t) + r\Delta tP(t) + k\Delta t + \mathcal{O}(\Delta t^2)
$$
$$
\Longleftrightarrow 
P(t + \Delta t) \approx P(t) + r\Delta tP(t) + k\Delta t
$$

From this point *using the hint from above* we can substitute $t=n\Delta t$ to get:

$$
P((n+1)\Delta t) \approx P(n\Delta t) + r\Delta t P(n\Delta t) + k\Delta t
$$

Then letting $P_{n}=P(n\Delta t)$ gives us the difference equation 

$$
P_{n+1} = P_{n} + r\Delta t P_{n} + k\Delta t = (1+r\Delta t)P_{n} + k\Delta t, ~ n\geq 0.
$$

Since $P_{n} = (1+r\Delta t)^{n}\left( P_{0} + \frac{k}{r}\right) - \frac{k}{r}$ we now have:

$$
P_{n+1} = (1+r\Delta t)^{n+1}\left( P_{0} + \frac{k}{r}\right) - \frac{k}{r}
$$

Using (previously proved) $(1 + r\delta t)^n \approx e^{rn\delta t}$ we can say:

$$
P_{n+1} = e^{(1+n)r\Delta t}\left( P_{0} + \frac{k}{r}\right) - \frac{k}{r}
$$

Since $e^{r\Delta t} \rightarrow 1$ as $\Delta t \rightarrow 0$ and $n\Delta t \rightarrow nt$ as $\Delta t \rightarrow 0$ and $n \rightarrow \infty$

$$
e^{(1+n)r\Delta t} = e^{rt}
$$

And therefore:

$$
P = e^{rt}\left( P_{0} + \frac{k}{r}\right) - \frac{k}{r}
$$
